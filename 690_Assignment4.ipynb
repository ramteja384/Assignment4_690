Dataframe with required 9 variables

```python
df = data[['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']]
print(df.head())
```

*Definitions of the 9 Variables/Columns
**UNTID:- UNITID is a Unique Identification number assigned to postsecondary institutions.
**INSTNM:- INSTNM represents the name of the institution.
**CITY:- CITY includes all the cities in which each Institution is located.
**STABBR:- Represents the location of the institution.
**ZIP:- Includes the area code or Zip code of that particular institution.
**ADM_RATE:- ADM_RATE includes the admissions rate at each campus.
**UGDS:- UGDS includes the total enrollment of Undergraduate Degree-Seeking students.
**TUITIONFEE_IN:- Includes the in-state tutiton fee details of the students
**MN_EARN_WNE_P:- MN_EARN_WNE_P earnings are for the institutional aggregate of all federally aided students who enroll in an institution each year.

Find and displaying UMBC

```python
df1 = pd.DataFrame(columns=['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10'])
for index in range(len(df)):
  if df.loc[index,'INSTNM'] == 'University of Maryland-Baltimore County':
    print(df.loc[index])
    df1 = df1.append(df.loc[index])
```

Finding UMD 

```python
for index in range(len(df)):
  if df.loc[index,'INSTNM'] == 'University of Maryland-College Park':
    print(df.loc[index])
    df1 = df1.append(df.loc[index])
```

Finding Loyola

```python
for index in range(len(df)):
  if df.loc[index,'INSTNM'] == 'Loyola University Maryland':
    print(df.loc[index])
    df1 = df1.append(df.loc[index])
```

Find and displaying Johns Hopkins

```python
for index in range(len(df)):
  if df.loc[index,'INSTNM'] == 'Johns Hopkins University':
    print(df.loc[index])
    df1 = df1.append(df.loc[index])
```

Calculating the min, max, and average in-state tution of the above four institutions.

```python
print(df1['TUITIONFEE_IN'].min())
print(df1['TUITIONFEE_IN'].max())
print(df1['TUITIONFEE_IN'].mean())
```

Colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia

```python
result=pd.DataFrame(columns=['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10'])
states = ['VA','MD','WA','PA','DE']
for index in range(len(df)):
  if df.loc[index,'STABBR'] in states:
    result = result.append(df.loc[index])
```
